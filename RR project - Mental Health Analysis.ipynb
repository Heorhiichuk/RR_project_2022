```python
# Changed all the YES/NO into 1/0
df_d_new = df_d.replace({'Yes': 1,'No': 0})

# Changed the currency of the Income (1 U.S. dollar = 4.47753843 Polish zloty)
df_d_new['Household Income'] = df_d_new['Household Income']/4.47753843
df_d_new['Annual income (including any social welfare programs) in USD'] = df_d_new['Annual income (including any social welfare programs) in USD']/4.47753843
df_d_new['Annual income from social welfare programs'] = df_d_new['Annual income from social welfare programs']/4.47753843
```

```python
# Check the interval in column 'Household Income'
df_v['Household Income'].nunique()
# $0-$9,999
# $10,000-$24,999
# $25,000-$49,999
# $50,000-$74,999
# $75,000-$99,999
# $100,000-$124,999
# $125,000-$149,999
# $150,000-$174,999
# $175,000-$199,999
# $200,000+
# Prefer not to answer
```

```python
# Change the integer into interval string same as Clean Data
All_Conditions = [
    (df_d_new['Household Income'] >= 200000),
    (df_d_new['Household Income'] >= 175000) & (df_d_new['Household Income'] <= 199999),
    (df_d_new['Household Income'] >= 150000) & (df_d_new['Household Income'] <= 174999),
    (df_d_new['Household Income'] >= 125000) & (df_d_new['Household Income'] <= 149999),
    (df_d_new['Household Income'] >= 100000) & (df_d_new['Household Income'] <= 124999),
    (df_d_new['Household Income'] >= 75000) & (df_d_new['Household Income'] <= 99999),
    (df_d_new['Household Income'] >= 50000) & (df_d_new['Household Income'] <= 74999),
    (df_d_new['Household Income'] >= 25000) & (df_d_new['Household Income'] <= 49999),
    (df_d_new['Household Income'] >= 10000) & (df_d_new['Household Income'] <= 24999),
    (df_d_new['Household Income'] >= 0) & (df_d_new['Household Income'] <= 9999)
]
All_Categories = ['$200,000+','$175,000-$199,999', '$150,000-$174,999', 
                  '$125,000-$149,999', '$100,000-$124,999', 
                  '$75,000-$99,999' ,'$50,000-$74,999', 
                  '$25,000-$49,999', '$10,000-$24,999', 
                  '$0-$9,999'
]
df_d_new['Household Income'] = np.select(All_Conditions, All_Categories)

```
